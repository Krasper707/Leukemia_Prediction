#Feature Engineering

```python
from sklearn.preprocessing import LabelEncoder
```

- Categorization of BMI helps to identify major groups in the data.

```python
def categorize_bmi(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 25:
        return "Normal"
    elif 25 <= bmi < 30:
        return "Overweight"
    else:
        return "Obese"

data["BMI_Category"] = data["BMI"].apply(categorize_bmi)
```

- WBC/RBC Ratio, Hemoglobin/RBC ratio,etc are a good measure of Leukemia


```python
data["WBC_RBC_Ratio"] = data["WBC_Count"] / data["RBC_Count"]

data["Hemoglobin_RBC_Ratio"] = data["Hemoglobin_Level"] / data["RBC_Count"]

data["Platelet_WBC_Ratio"] = data["Platelet_Count"] / data["WBC_Count"]

data["BoneMarrowBlasts_WBC_Ratio"] = data["Bone_Marrow_Blasts"] / data["WBC_Count"]
```

```python
data["Anemia_Risk"] = ((data["Hemoglobin_Level"] < 12) & (data["RBC_Count"] < 4.5)).astype(int)
data["Weak_Immune_System"] = ((data["WBC_Count"] < 4000) &
                            ((data["Smoking_Status"] == "Yes") | (data["Alcohol_Consumption"] == "Yes"))).astype(int)
data["Leukemia_Risk_Score"] = (
    (data["Bone_Marrow_Blasts"] > 50).astype(int) +
    (data["WBC_Count"] > 10000).astype(int) +
    (data["Genetic_Mutation"] == "Yes").astype(int)
)
```

```python
data["Age_BMI_Interaction"] = data["Age"] * data["BMI"]
data["Age_WBC_Interaction"] = data["Age"] * data["WBC_Count"]

data["Age_Group"] = pd.cut(data["Age"], bins=[0, 30, 50, 100], labels=["Young", "Middle_Aged", "Old"])
```

```python
data["High_Genetic_Risk"] = ((data["Family_History"] == "Yes") & (data["Genetic_Mutation"] == "Yes")).astype(int)

data["Lifestyle_Risk"] = ((data["Smoking_Status"] == "Yes").astype(int) + (data["Alcohol_Consumption"] == "Yes").astype(int))
```

```python
encoder = LabelEncoder()

categorical_columns = ['Gender','Genetic_Mutation', 'Family_History', 'Smoking_Status'
                       , 'Alcohol_Consumption', 'Radiation_Exposure', 'Infection_History',
                        'Chronic_Illness', 'Immune_Disorders', 'Ethnicity', 'Socioeconomic_Status',
                       'Urban_Rural', 'Leukemia_Status', 'BMI_Category', 'Age_Group' ]

for column in categorical_columns:
    data[column] = encoder.fit_transform(data[column])

```

#To counter imbalanced dataset, SMOTE can be used.
- Reference: [Link](https://machinelearningmastery.com/smote-oversampling-for-imbalanced-classification/)

```python
from imblearn.over_sampling import SMOTE
x = data.drop(["Leukemia_Status", "Country"], axis=1)
y=data[["Leukemia_Status"]]
smote = SMOTE(random_state=42)
x_resampled, y_resampled = smote.fit_resample(x, y)
y_resampled.value_counts()
```

#Model selection,Training and testing.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

x_train, x_test, y_train, y_test = train_test_split(x_resampled, y_resampled, test_size=.2, random_state = 42)
```

```python
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
tahmin = dt.predict(x_test)

accuracy_score(y_test, tahmin)
```

```python
print(classification_report(y_test, tahmin))
```

```python
sns.heatmap(confusion_matrix(y_test, tahmin), annot=True)
```

```python
rf = RandomForestClassifier()

rf.fit(x_train, y_train)
tahmin = rf.predict(x_test)

accuracy_score(y_test, tahmin)
```

```python
print(classification_report(y_test, tahmin))
```

```python
sns.heatmap(confusion_matrix(y_test, tahmin), annot=True)
```

###RandomForest is the best classifier amongst all.
